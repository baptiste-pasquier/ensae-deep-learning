```python
import functools
from pathlib import Path

import torch
import torchvision.transforms as transforms
from torchvision.datasets import EMNIST

from ensae_deep_learning.sampling import euler_maruyama_sampler, ode_sampler, pc_sampler
from ensae_deep_learning.sde import ScoreNet, diffusion_coeff, marginal_prob_std
from ensae_deep_learning.utils import (
    hflip,
    plot_dataset,
    plot_loss_history,
    plot_samples,
    print_sde_dim,
    rotate_90,
    run_training,
    summary_model,
)

device = "cuda"
```

```python
sigma = 25.0
marginal_prob_std_fn = functools.partial(marginal_prob_std, sigma=sigma)
diffusion_coeff_fn = functools.partial(diffusion_coeff, sigma=sigma)
```

## 1. Data preprocessing

```python
dataset_original = EMNIST(
    "data/", split="balanced", transform=transforms.ToTensor(), download=True
)
```

```python
config = {
    "model_name": "EMNIST_2",
    "image_size": (25, 25),
    "in_channels": 1,
    "channels": (32, 64, 128, 256),
    "embed_dim": 256,
    "n_epochs": 50,
    "lr": 1e-4,
    "batch_size": 64,
}

mean = torch.tensor([0.1751])
std = torch.tensor([0.3332])

normalize = transforms.Normalize(mean, std)
unnormalize = transforms.Normalize((-mean / std), (1.0 / std))

transformation = transforms.Compose(
    [
        transforms.Lambda(rotate_90),
        transforms.Lambda(hflip),
        transforms.CenterCrop(config["image_size"]),
        transforms.RandomAffine(degrees=20, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        transforms.ToTensor(),
        normalize,
    ]
)
dataset_transformed = EMNIST(
    "data/", split="balanced", transform=transformation, download=True
)
```

```python
save_dir = Path("models", config["model_name"])
save_dir.mkdir(exist_ok=True)

plot_dataset(dataset_original, save_dir, "dataset_original.pdf")
plot_dataset(dataset_transformed, save_dir, "dataset_transformed.pdf", unnormalize)
```

## 2. Training

```python
score_model = torch.nn.DataParallel(
    ScoreNet(
        in_channels=config["in_channels"],
        marginal_prob_std=marginal_prob_std_fn,
        channels=config["channels"],
        embed_dim=config["embed_dim"],
    )
)
score_model = score_model.to(device)

summary_model(score_model, config)
```

```python
print_sde_dim(config)
```

```python
loss_history = run_training(
    score_model,
    dataset_transformed,
    config,
    marginal_prob_std_fn,
    save_dir,
    device,
)
```

```python
plot_loss_history(loss_history, save_dir)
```

## 3. Sampling

```python
# Load the pre-trained checkpoint from disk.
ckpt = torch.load(Path(save_dir, "model.pth"), map_location=device)
score_model.load_state_dict(ckpt)

sample_batch_size = 36
for sampler in [
    euler_maruyama_sampler,
    pc_sampler,
    ode_sampler,
]:
    # Generate samples using the specified sampler.
    samples = sampler(
        score_model,
        marginal_prob_std_fn,
        diffusion_coeff_fn,
        config,
        sample_batch_size,
        device=device,
    )
    samples = unnormalize(samples)

    plot_samples(samples, sampler, save_dir)
```
